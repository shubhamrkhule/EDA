Plotting in EDA consists of Histograms, Box plot, Scatter plot and many more.

### as you can see there consists a lot of missing values in our dataset , we will use the mean technique in order to fill them below

```python
#using mean to impute all the missing values
missing_col = ['incidenceRate','PctSomeCol18_24','PctEmployed16_Over','PctPrivateCoverageAlone','PctWhite']
for i in missing_col:
 data.loc[data.loc[:,i].isnull(),i]=data.loc[:,i].mean()
```
